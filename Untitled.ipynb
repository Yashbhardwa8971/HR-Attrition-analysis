```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
pd.set_option('display.max_columns', None)
```

```python
import os

os.makedirs("images", exist_ok=True)
```

```python
df = pd.read_csv("HR_Employee_Attrition.csv")

print(df.shape)
df.head()
```

```python
print(df.info())

print("\nMissing Values")
print(df.isnull().sum())

print("\nDuplicate Rows")
print(df.duplicated().sum())
```

```python
df.describe()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x='Attrition', data=df)

plt.title('Employee Attrition Distribution')

plt.savefig("images/attrition_distribution.png")
plt.savefig("attrition_distribution.png")
plt.show()
```

```python
attrition_percent = df['Attrition'].value_counts(normalize=True)*100

print(attrition_percent)
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Gender',
    hue='Attrition',
    data=df
)

plt.title("Attrition by Gender")

plt.savefig("images/gender_attrition.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Department',
    hue='Attrition',
    data=df
)

plt.xticks(rotation=20)

plt.title("Department Wise Attrition")

plt.savefig("images/department_attrition.png")
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(
    y='JobRole',
    hue='Attrition',
    data=df
)

plt.title("Attrition by Job Role")

plt.savefig("images/jobrole_attrition.png")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    x='Attrition',
    y='MonthlyIncome',
    data=df
)

plt.title("Monthly Income vs Attrition")

plt.savefig("images/income_attrition.png")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    x='Department',
    y='MonthlyIncome',
    data=df
)

sns.stripplot(
    x='Department',
    y='MonthlyIncome',
    data=df,
    color='black',
    alpha=0.3
)

plt.title('Monthly Income Distribution by Department')

plt.savefig("images/monthly_income_distribution_department.png")

plt.show()
```

```python
dept_salary = (
    df.groupby('Department')['MonthlyIncome']
    .mean()
    .sort_values(ascending=False)
)

print(dept_salary)
```

```python
plt.figure(figsize=(8,5))

dept_salary.plot(kind='bar')

plt.title('Average Monthly Income by Department')
plt.ylabel('Average Monthly Income')

plt.savefig("images/avg_salary_department.png")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    x='Department',
    y='MonthlyIncome',
    hue='Attrition',
    data=df
)

plt.title('Monthly Income by Department and Attrition')

plt.savefig("images/income_department_attrition.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x='Age',
    hue='Attrition',
    kde=True
)

plt.title("Age Distribution")

plt.savefig("images/age_distribution.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Attrition',
    y='YearsAtCompany',
    data=df
)

plt.title("Years at Company vs Attrition")

plt.savefig("images/years_company.png")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='OverTime',
    hue='Attrition',
    data=df
)

plt.title("Overtime Impact on Attrition")

plt.savefig("images/overtime_attrition.png")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(18,10))

sns.heatmap(
    numeric_df.corr(),
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.savefig("images/correlation_heatmap.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='JobSatisfaction',
    hue='Attrition',
    data=df
)

plt.title("Job Satisfaction vs Attrition")

plt.savefig("images/job_satisfaction.png")
plt.show()
```
